### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt

# Data for the plots
x_line = [0, 1, 2, 3, 4, 5]
y_line = [0, 1, 4, 9, 16, 25]

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart in the first subplot
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter)
for i, j in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f'({i}, {j})', (i, j))
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('x values')
axs[1, 0].set_ylabel('y values')

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
